```python
import pandas as pd
```

```python
ccba_df = pd.read_csv('/content/drive/MyDrive/訓練資料集_first/public_train_x_ccba_full_hashed.csv')
ccba_df = ccba_df.groupby('cust_id').mean().drop(columns=['byymm'])
ccba_df
```

```python
cdtx_df = pd.read_csv('/content/drive/MyDrive/訓練資料集_first/public_train_x_cdtx0001_full_hashed.csv', index_col='cust_id')
cdtx_df.country
idx_freq_dict = {}
for idx in cdtx_df.index.values.tolist():
  if idx not in idx_freq_dict:
    idx_freq_dict[idx] = 1
  else:
    idx_freq_dict[idx] += 1
idx_list = list(dict.fromkeys(cdtx_df.index.values.tolist()))
cust_id_result = []
country_result = []
cur_type_result = []
amt_result = []
freq_result = []
for idx in idx_list:
  if idx_freq_dict[idx] > 1:
    cust_id_result.append(idx)
    country_result.append(cdtx_df.loc[idx]['country'].mode().iloc[0])
    cur_type_result.append(cdtx_df.loc[idx]['cur_type'].mode().iloc[0])
    amt_result.append(cdtx_df.loc[idx]['amt'].mean())
    freq_result.append(idx_freq_dict[idx])
  else:
    cust_id_result.append(idx)
    country_result.append(cdtx_df.loc[idx]['country'])
    cur_type_result.append(cdtx_df.loc[idx]['cur_type'])
    amt_result.append(cdtx_df.loc[idx]['amt'].mean())
    freq_result.append(idx_freq_dict[idx])

data = {'cust_id':cust_id_result,'country':country_result,'cur_type':cur_type_result,'amt':amt_result,'con_freq':freq_result}
cdtx_df = pd.DataFrame(data)
cdtx_df
#cdtx_df.loc['4220624de6d38bd4b41d1c32932e03f3c0e85b1b08b7e399df191feb73d3a4c1']
#5b833abe374782cf4a49db8e89918161c6f9c4e4ec7618752aeb592211dec4e1
#print(cdtx_df.loc['4220624de6d38bd4b41d1c32932e03f3c0e85b1b08b7e399df191feb73d3a4c1']['country'].mode())
#print(len(cdtx_df.loc['5b833abe374782cf4a49db8e89918161c6f9c4e4ec7618752aeb592211dec4e1']))
#idx_freq_dict['5b833abe374782cf4a49db8e89918161c6f9c4e4ec7618752aeb592211dec4e1']
```

```python
remit1_df = pd.read_csv('/content/drive/MyDrive/訓練資料集_first/public_train_x_remit1_full_hashed.csv', index_col='cust_id')
remit1_df.drop(columns=['trans_date'])

idx_freq_dict = {}
for idx in remit1_df.index.values.tolist():
  if idx not in idx_freq_dict:
    idx_freq_dict[idx] = 1
  else:
    idx_freq_dict[idx] += 1
idx_list = list(dict.fromkeys(remit1_df.index.values.tolist()))
cust_id_result = []
trans_no_result = []
trade_amount_usd_result = []
freq_result = []
for idx in idx_list:
  if idx_freq_dict[idx] > 1:
    cust_id_result.append(idx)
    trans_no_result.append(remit1_df.loc[idx]['trans_no'].mode().iloc[0])
    trade_amount_usd_result.append(remit1_df.loc[idx]['trade_amount_usd'].mean())
    freq_result.append(idx_freq_dict[idx])
  else:
    cust_id_result.append(idx)
    trans_no_result.append(remit1_df.loc[idx]['trans_no'])
    trade_amount_usd_result.append(remit1_df.loc[idx]['trade_amount_usd'].mean())
    freq_result.append(idx_freq_dict[idx])

data = {'cust_id':cust_id_result,'trans_no':trans_no_result,'trade_amount_usd':trade_amount_usd_result,'remit_freq':freq_result}
remit1_df = pd.DataFrame(data)
remit1_df
```

```python
dp_df = pd.read_csv('/content/drive/MyDrive/訓練資料集_first/public_train_x_dp_full_hashed.csv', index_col='cust_id')

idx_freq_dict = {}
for idx in dp_df.index.values.tolist():
  if idx not in idx_freq_dict:
    idx_freq_dict[idx] = 1
  else:
    idx_freq_dict[idx] += 1
idx_list = list(dict.fromkeys(dp_df.index.values.tolist()))
cust_id_result = []
debit_credit_result = []
tx_type_result = []
tx_amt_result = []
exchg_rate_result = []
info_asset_code_result = []
tx_type_x_info_asset_code_result = []
fiscTxId_result = []
txbranch_result = []
cross_bank_result = []
ATM_result = []
freq_result = []

for idx in idx_list:
  if idx_freq_dict[idx] > 1:
    cust_id_result.append(idx)
    debit_credit_result.append(dp_df.loc[idx]['debit_credit'].mode().iloc[0])
    tx_type_result.append(dp_df.loc[idx]['tx_type'].mode().iloc[0])
    exchg_rate_result.append(dp_df.loc[idx]['exchg_rate'].mode().iloc[0])
    info_asset_code_result.append(dp_df.loc[idx]['info_asset_code'].mode().iloc[0])
    fiscTxId_result.append(dp_df.loc[idx]['fiscTxId'].mode(dropna=False).iloc[0])
    txbranch_result.append(dp_df.loc[idx]['txbranch'].mode(dropna=False).iloc[0])
    cross_bank_result.append(dp_df.loc[idx]['cross_bank'].mode().iloc[0])
    ATM_result.append(dp_df.loc[idx]['ATM'].mode().iloc[0])
    tx_amt_result.append(dp_df.loc[idx]['tx_amt'].mean())
    tx_type_x_info_asset_code_result.append({True:1,False:0}[(dp_df.loc[idx]['tx_type']*dp_df.loc[idx]['info_asset_code']).mode().iloc[0]==12])
    freq_result.append(idx_freq_dict[idx])
  else:
    cust_id_result.append(idx)
    debit_credit_result.append(dp_df.loc[idx]['debit_credit'])
    tx_type_result.append(dp_df.loc[idx]['tx_type'])
    exchg_rate_result.append(dp_df.loc[idx]['exchg_rate'])
    info_asset_code_result.append(dp_df.loc[idx]['info_asset_code'])
    fiscTxId_result.append(dp_df.loc[idx]['fiscTxId'])
    txbranch_result.append(dp_df.loc[idx]['txbranch'])
    cross_bank_result.append(dp_df.loc[idx]['cross_bank'])
    ATM_result.append(dp_df.loc[idx]['ATM'])
    tx_amt_result.append(dp_df.loc[idx]['tx_amt'].mean())
    tx_type_x_info_asset_code_result.append({True:1,False:0}[(dp_df.loc[idx]['tx_type']*dp_df.loc[idx]['info_asset_code'])==12])
    freq_result.append(idx_freq_dict[idx])

  
data = {'cust_id':cust_id_result,'debit_credit':debit_credit_result,'tx_type':tx_type_result,'exchg_rate':exchg_rate_result, 'info_asset_code':info_asset_code_result, \
        'fiscTxId':fiscTxId_result, 'txbranch':txbranch_result, 'cross_bank':cross_bank_result, 'ATM':ATM_result, 'tx_amt':tx_amt_result, \
        'tx_type_x_info_asset_code_result':tx_type_x_info_asset_code_result,'trans_freq':freq_result}
dp_df = pd.DataFrame(data)
dp_df
 
```
